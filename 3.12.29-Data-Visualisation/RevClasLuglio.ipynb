```python
import plotly.express as px
```

```python
df = px.data.iris()
df.head()
```

```python
fig = px.scatter(df, x = 'petal_length', y='petal_width', color='species')
fig.update_layout(width=800, height=500)
```

```python
df1= px.data.gapminder()
df1.head()
```

```python
fig1 = px.scatter(df1, x = 'gdpPercap', y='lifeExp', animation_frame='year', animation_group= 'country',color='continent', 
                  hover_name ='country',size = 'pop', 
                  log_x= True, size_max=55, range_x=[200,100000], range_y=[25,90])

fig1.update_layout(width=800, height=500)
fig1.show()
```
